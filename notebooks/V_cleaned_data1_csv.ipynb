```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
df = pd.read_csv('../data/processed/ml_ready_data1.csv')

# Basic info
print(df.info())
print(df.describe())

# Check target distribution
plt.figure(figsize=(10, 5))
sns.histplot(df['universityRankingNum'], bins=50, kde=True)
plt.title('Distribution of University Ranking')
plt.xlabel('University Ranking Number')
plt.ylabel('Count')
plt.show()

# Check for missing values
print("Missing values per column:")
print(df.isnull().sum())

# Check feature correlations
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, fmt=".2f")
plt.title("Feature Correlation Matrix")
plt.show()

# Check distributions of key numeric features
numeric_features = ['ieltsMarks', 'toefl_ibt', 'minimumGPA', 'tuitionFeeUSD']
for feat in numeric_features:
    plt.figure(figsize=(8, 4))
    sns.histplot(df[feat], bins=50, kde=True)
    plt.title(f'Distribution of {feat}')
    plt.show()

# Check presence of country and course level one-hot columns
print("Sample columns containing 'country_' and 'courseLevelSimplified_':")
print([col for col in df.columns if col.startswith('country_')][:10])
print([col for col in df.columns if col.startswith('courseLevelSimplified_')][:10])
```
